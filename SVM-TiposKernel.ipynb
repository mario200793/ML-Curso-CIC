## Diferentes tipos de Kernel

```python
%matplotlib inline
import numpy as np
import pandas as pd
import math 
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
X = np.c_[(.4, -.7),
          (-1.5, -1),
          (-1.4, -.9),
          (-1.3, -1.2),
          (-1.1, -.2),
          (-1.2, -.4),
          (-.5, 1.2),
          (-1.5, 2.1),
          (1, 1),
          (1.3, .8),
          (1.2, .5),
          (.2, -2),
          (.5, -2.4),
          (.2, -2.3),
          (0, -2.7),
          (1.3, 2.1)].T
```

```python
X
```

```python
Y = [0] * 8 + [1] * 8
```

```python
Y
```

```python
plt.figure(1,figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Accent, #Visualizamos ltodos los datos
                edgecolors='k')
plt.axis('tight')
x_min = -3
x_max = 3
y_min = -3
y_max = 3
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```

```python
def plot_svmKernels(model):
    plt.figure(1,figsize=(10, 5))
    plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],zorder=10,
                facecolors='none', linewidths=4 ,edgecolors='r')  #Visualizamos los vectores de soporte
    plt.scatter(X[:, 0], X[:, 1], c=Y,zorder=10, cmap=plt.cm.Accent, #Visualizamos ltodos los datos
                edgecolors='k')
    plt.axis('tight')
    x_min = -3
    x_max = 3
    y_min = -3
    y_max = 3
    
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    Z = model.decision_function(np.c_[XX.ravel(), YY.ravel()])
    
    Z = Z.reshape(XX.shape)
    plt.figure(1, figsize=(10, 5))
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Pastel1)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])
                
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)


    plt.show()
```

Kernel Lineal

```python
clf = svm.SVC(kernel='linear')
clf.fit(X, Y)
plot_svmKernels(clf)
```

Kernel RBF (Radial basis function)

```python
clf = svm.SVC(kernel='rbf')
clf.fit(X, Y)
plot_svmKernels(clf)
```

Kernel Polinomial

```python
clf = svm.SVC(kernel='poly')
clf.fit(X, Y)
plot_svmKernels(clf)
```
